```python
import numpy as np
```

```python
"""
Introduction to Broadcasting
"""

# Broadcasting: 2D + 1D arrays

x = np.ones((3, 3))
print(x)
y = np.arange(3)
print(y)
print(x + y)
print('\n')

x1 = np.zeros((3, 3))
print(x1)
y1 = np.arange(3)
print(y1)
print(x1 + y1)
```

```python
# Broadcast both arrays

a = np.arange(3)
print(a)

a = a[:, np.newaxis]
print(a)

b = np.arange(3)
print(b)

print(a+b)
```

```python
"""
Broadcasting Rules
"""

a = np.ones((2, 3))
print(a)
print(a.shape, '\n')

b = np.arange(3)
print(b)
print(b.shape, '\n')

print(a+b)
```

```python
# Broadcast rule ex) boadcast both arrays

a = np.arange(3)
print(a)
a = a[:, np.newaxis]
print(a)
print(a.shape)

b = np.arange(3)
print(b.shape)

print(a+b)
```

```python
# Broadcast rule ex) Incompatible case

a = np.ones((3, 2))
print(a.shape)
b = np.arange(3)
print(b.shape)

print(a+b)
```

```python
"""
Comparisons, Masks, and Boolean Logic
"""
x = np.array([1, 2, 3, 4, 5])
print(x)

print(x < 3)
print(x >= 3)
print(x != 3)
print(x == 3)
```

```python
import numpy as np
import pandas as pd

# use pandas to extract rainfall inches as a NumPy array
rainfall = pd.read_csv('seattle2014.csv')['PRCP'].values
inches = rainfall / 254.0 # 1/10mm -> inches
inches.shape
print(inches)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # set plot styles
%config InlineBackend.figure_format = 'svg'
```

```python
plt.hist(inches, 40);
```

```python
print("Number of rainy day: ", np.count_nonzero(inches > 0), "days")
# == np.sum(inches > 0)
rainy_day = inches[inches > 0]
print("Averge precipitation on a rainy day: ", np.mean(rainy_day))
print("Number of days with more than 0.5 inches of rain: ", np.sum(rainy_day > 0.5), "days")
```

```python
# bool masking for multi-diemnsional arrays

np.random.seed(0)
x = np.random.randint(10, size=(3, 4))
print(x)

print(x < 6)
```

```python
"""
Working with Boolean Arrays
"""

print(x)
print(np.count_nonzero(x < 6))
print(np.sum(x < 6))

print(np.any(x > 8))
print(np.any(x > 9))
print(np.all(x > 1))
```

```python
# Masking operations

print(x < 5)
print(x[x < 5])

print(np.max(x[x < 5]))
print(np.min(x[x < 5]))
print(np.mean(x[x < 5]))
print(np.std(x[x < 5]))
```

```python
"""
Fancy Indexing
"""

np.random.seed(1)
x = np.random.randint(100, size=10)
print(x)

print(x[2], x[4], x[6])

ind = [2, 4, 6]
print(x[ind])
```

```python
# For n-D arrays

np.random.seed(2)
x = np.random.randint(10, size=(3, 5))
print(x)

row = [0, 1, 2]
col = [3, 1, 2]

x[row, col]
```

```python
# For n-D arrays with broadcasting

print(x)
row = np.array([0, 1, 2])
col = np.array([2, 1, 4])

print(x[row[:, np.newaxis], col])
```

```python
"""
Combined Indexing
"""
x = np.arange(1, 13).reshape(3, 4)
print(x)

print(x[2, [2, 0, 1]])
print(x[1:, [2, 0, 1]])
```
