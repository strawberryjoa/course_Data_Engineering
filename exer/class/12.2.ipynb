```python
"""
Exercise 1: Search '메타버스' news on Naver
"""

# HTTP Request
import requests

url1 = 'https://search.naver.com/search.naver?where=nexearch&sm=top_sug.pre&fbm=1&acr=1&acq=apxkqj&qdt=0&ie=utf8&query=%EB%A9%94%ED%83%80%EB%B2%84%EC%8A%A4'
res = requests.get(url1) # HTTP GET Request

html_doc = res.text # get texts from the HTML response result

print(html_doc)
```

```python
"""
User BeautifulSoup Parser
"""

from bs4 import BeautifulSoup
# html, xml, lxml
# Create a parser
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify()) # structuring the tags
```

```python
# Search all the <a> tags

soup.find_all('a')
```

```python
# Extract news information

news = soup.find_all("a", {"class":"news_tit"})
print(news)
```
